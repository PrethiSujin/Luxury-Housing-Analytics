# handling outliers

```python
Q1 = df['Ticket_Price_Cr'].quantile(0.25)
Q3 = df['Ticket_Price_Cr'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].clip(lower, upper)
```

```python
import pyodbc

conn = pyodbc.connect(
     "Driver={SQL Server};"
        "Server=LAPTOP-06C3FHVP\SQLEXPRESS;"
        "Database=LuxuryHousingDB;"
        "Trusted_Connection=yes;"
)

cursor = conn.cursor()


data = df.values.tolist()
    

insert_query = """
INSERT INTO Luxury_Housing(
            property_id,
            micro_market,
            project_name,
            builder,
            unit_size_sqft,
            configuration,
            ticket_price_cr,
            transaction_type,
            buyer_type,
            purchase_quarter,
            connectivity_score,
            amenity_score,
            possession_status,
            sales_channel,
            nri_buyer,
            locality_infra_score,
            avg_traffic_time_min,
            buyer_comments,
            booking_flag,
            booking_status,
            booking_success,
            price_rupees,
            price_per_sqft)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""

cursor.executemany(insert_query, data)
conn.commit()

print(f"✅ {cursor.rowcount} records inserted successfully!")
cursor.execute("SELECT * FROM Luxury_Housing")
for row in cursor.fetchall():
    print(row)

cursor.close()
conn.close()
```
